```python
import html
```

```python
s = html.unescape(sample['text']) 
```

```python
def myfilter(s):
    s = s.replace('<strong>','')
    s = s.replace('</strong>','')
    return s
```

```python
s = myfilter(s)
```

```python
print(s)
```

```python
from pyltp import SentenceSplitter
```

```python
sents = SentenceSplitter.split(s)  # 分句
print('\n'.join(list(sents)))
```

```python
from pyltp import Segmentor
```

```python
segmentor = Segmentor()
segmentor.load("./ltp_data_v3.4.0/cws.model")
words = segmentor.segment(s)
print("|".join(words))
segmentor.release()
```

```python
from pyltp import Postagger
pos_model_path = './ltp_data_v3.4.0/pos.model'
```

```python

postagger = Postagger() # 初始化实例
postagger.load(pos_model_path)  # 加载模型

words = ['元芳', '你', '怎么', '看']  # 分词结果
postags = postagger.postag(words)  # 词性标注

print('\n'.join(postags))
postagger.release()  # 释放模型
```

```python
from pyltp import NamedEntityRecognizer
ner_model_path = './ltp_data_v3.4.0/ner.model'
```

```python
recognizer = NamedEntityRecognizer() # 初始化实例
recognizer.load(ner_model_path)  # 加载模型

words = ['元芳', '你', '怎么', '看']
postags = ['nh', 'r', 'r', 'v']
netags = recognizer.recognize(words, postags)  # 命名实体识别

print('\n'.join(netags))
recognizer.release()  # 释放模型
```
