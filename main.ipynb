```python
import html
```

```python
s = html.unescape(sample['text']) 
```

```python
def myfilter(s):
    s = s.replace('<strong>','')
    s = s.replace('</strong>','')
    return s
```

```python
s = myfilter(s)
```

```python
print(s)
```

```python
from pyltp import SentenceSplitter
```

```python
sents = SentenceSplitter.split(s)  # åˆ†å¥
print('\n'.join(list(sents)))
```

```python
from pyltp import Segmentor
```

```python
segmentor = Segmentor()
segmentor.load("./ltp_data_v3.4.0/cws.model")
words = segmentor.segment(s)
print("|".join(words))
segmentor.release()
```

```python
from pyltp import Postagger
pos_model_path = './ltp_data_v3.4.0/pos.model'
```

```python

postagger = Postagger() # åˆå§‹åŒ–å®ä¾‹
postagger.load(pos_model_path)  # åŠ è½½æ¨¡å‹

words = ['å…ƒèŠ³', 'ä½ ', 'æ€ä¹ˆ', 'çœ‹']  # åˆ†è¯ç»“æœ
postags = postagger.postag(words)  # è¯æ€§æ ‡æ³¨

print('\n'.join(postags))
postagger.release()  # é‡Šæ”¾æ¨¡å‹
```

```python
from pyltp import NamedEntityRecognizer
ner_model_path = './ltp_data_v3.4.0/ner.model'
```

```python
recognizer = NamedEntityRecognizer() # åˆå§‹åŒ–å®ä¾‹
recognizer.load(ner_model_path)  # åŠ è½½æ¨¡å‹

words = ['å…ƒèŠ³', 'ä½ ', 'æ€ä¹ˆ', 'çœ‹']
postags = ['nh', 'r', 'r', 'v']
netags = recognizer.recognize(words, postags)  # å‘½åå®ä½“è¯†åˆ«

print('\n'.join(netags))
recognizer.release()  # é‡Šæ”¾æ¨¡å‹
```
